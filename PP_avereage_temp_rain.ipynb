```python
import xarray as xr
import matplotlib.pyplot as plt
import geopandas as gpd
import glob
import numpy as np

# =========================================
# 1️⃣ Load Shapefile
# =========================================
shapefile_path = r'/content/drive/MyDrive/Research Works - Development/Work 2 - SPI SPEI/Eastern_Aus_Bunbury/Eastern_Aus_Bunbury.shp'
qld_shape = gpd.read_file(shapefile_path).to_crs("EPSG:4326")

# =========================================
# 2️⃣ Open ALL 33 NetCDF Files
# =========================================
nc_files = sorted(glob.glob(r'/content/drive/MyDrive/Research Works - Development/Work 2 - SPI SPEI/Data/raster/combined_climate/climate_*.nc'))

print(f"Total files found: {len(nc_files)}")

# Open multiple files as one dataset
ds = xr.open_mfdataset(nc_files, combine='by_coords')

# =========================================
# 3️⃣ Select Variables
# =========================================
monthly_rain = ds['monthly_rain']
max_temp = ds['max_temp']

# =========================================
# 4️⃣ Compute Mean Across ALL Time (33*12)
# =========================================
mean_rain = monthly_rain.mean(dim='time')
mean_temp = max_temp.mean(dim='time')

# =========================================
# 5️⃣ Print Overall Spatial Mean Values
# =========================================
print("Average monthly_rain (overall spatial mean):",
      float(mean_rain.mean().values))

print("Average max_temp (overall spatial mean):",
      float(mean_temp.mean().values))

# =========================================
# 6️⃣ Plot Results
# =========================================
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# --- Rainfall ---
mean_rain.plot(ax=axes[0], cmap='Blues', add_colorbar=True)
qld_shape.boundary.plot(ax=axes[0], color='black', linewidth=0.8)
axes[0].set_title("Average Monthly Rain (33 Years)")
axes[0].set_xlabel('')
axes[0].set_ylabel('')

# --- Temperature ---
mean_temp.plot(ax=axes[1], cmap='coolwarm', add_colorbar=True)
qld_shape.boundary.plot(ax=axes[1], color='black', linewidth=0.8)
axes[1].set_title("Average Max Temperature (33 Years)")
axes[1].set_xlabel('')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt

# Load shapefile
shape_path = "/content/drive/MyDrive/Research Works - Development/Work 2 - SPI SPEI/Eastern_Aus_Bunbury/Eastern_Aus_Bunbury.shp"
gdf_shape = gpd.read_file(shape_path)

# Load NetCDF
ds = xr.open_dataset("/content/drive/MyDrive/Research Works - Development/Work 2 - SPI SPEI/Data/raster/Farm Locations Test/EA/f2022.c2002.p2022.t2022_ea.nc")

# Convert raster to dataframe
df = ds["A_wheat_hat_ha"].to_dataframe().reset_index()

# Remove NaN values
df = df.dropna()

# Create GeoDataFrame from lat/lon
gdf_points = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df["lon"], df["lat"]),
    crs="EPSG:4326"
)

# Reproject if needed
gdf_points = gdf_points.to_crs(gdf_shape.crs)

# Plot
fig, ax = plt.subplots(figsize=(10, 8))

# Plot boundary
gdf_shape.plot(
    ax=ax,
    color="white",
    edgecolor="black",
    linewidth=1
)

# Plot points (same look as your CSV version)
gdf_points.plot(
    ax=ax,
    column="A_wheat_hat_ha",
    cmap="YlOrRd",
    markersize=1,
    legend=True,
    alpha=0.8
)

# ax.set_title("Wheat Production Distribution", fontsize=14)
ax.set_axis_off()

plt.show()
```

```python
import geopandas as gpd
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# Load shapefile
shape_path = "/content/drive/MyDrive/Research Works - Development/Work 2 - SPI SPEI/Eastern_Aus_Bunbury/Eastern_Aus_Bunbury.shp"
gdf_shape = gpd.read_file(shape_path)

# Load NetCDF
ds = xr.open_dataset("/content/drive/MyDrive/Research Works - Development/Work 2 - SPI SPEI/Data/raster/Farm Locations Test/EA/f2022.c2002.p2022.t2022_ea.nc")

# Convert raster to dataframe
df = ds["A_wheat_hat_ha"].to_dataframe().reset_index()
df = df.dropna()

# Create GeoDataFrame
gdf_points = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df["lon"], df["lat"]),
    crs="EPSG:4326"
)

# Reproject if needed
gdf_points = gdf_points.to_crs(gdf_shape.crs)

# ----------------------------------------
# Plot
# ----------------------------------------
fig, ax = plt.subplots(figsize=(10, 8))

# Plot boundary
gdf_shape.plot(
    ax=ax,
    color="white",
    edgecolor="black",
    linewidth=1
)

# Create normalization for colorbar
norm = mpl.colors.Normalize(
    vmin=gdf_points["A_wheat_hat_ha"].min(),
    vmax=gdf_points["A_wheat_hat_ha"].max()
)

# Scatter plot manually (better control)
sc = ax.scatter(
    gdf_points.geometry.x,
    gdf_points.geometry.y,
    c=gdf_points["A_wheat_hat_ha"],
    cmap="YlOrRd",
    s=1,
    alpha=0.8,
    norm=norm
)

# Create colorbar
cbar = plt.colorbar(sc, ax=ax)

# 🔥 Increase font sizes
cbar.ax.tick_params(labelsize=14)      # Tick numbers
cbar.set_label("Wheat Area Proportion", fontsize=16)  # Colorbar title

ax.set_axis_off()

plt.show()
```
